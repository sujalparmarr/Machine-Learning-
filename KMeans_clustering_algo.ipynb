# KMeans Clustering - Beginner Friendly Implementation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

### Step 1: Generate synthetic data using `make_blobs`

```python
# Generate data with 4 centers
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
X = pd.DataFrame(X, columns=['Feature_1', 'Feature_2'])
```

### Step 2: Visualize the data

```python
plt.scatter(X['Feature_1'], X['Feature_2'], s=30)
plt.title('Generated Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

### Step 3: Apply KMeans Clustering

```python
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

### Step 4: Visualize the Clusters with Centroids

```python
plt.scatter(X['Feature_1'], X['Feature_2'], c=y_kmeans, s=30, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=100, alpha=0.75, label='Centroids')
plt.title('KMeans Clustering Result')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
